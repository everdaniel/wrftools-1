#### This is a small demo of how to use wrftools in conjuction with numpy ans netCDF4. Make sure this document is in the same directory as the file wrf_testfile_in.nc. We'll start by importing the tools that we will need.

```python
from netCDF4 import Dataset
from wrftools.variables import *
from wrftools.interp import *
##for the end of the example...
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

#### Open the netCDF file with the data to be processed, and create a new netCDF file to write the processed data to.

```python
infile = Dataset( 'wrf_testfile_in.nc' )
outfile = Dataset( 'wrf_out.nc', 'w', format='NETCDF4' )

print infile.variables.keys()
```

#### The file contains 5 variables: XLONG, XLAT, P, PB, and T. We will be interpolating potential temperature onto pressure surfaces for this example. However, pressure and potential temperature must first be calculated from the data output by the WRF.

```python
## read in the variables
P_input = infile.variables['P'][:]
PB_input = infile.variables['PB'][:]

T_input = infile.variables['T'][:]
## calculate pressure and potentia temperature
PRES_input = wrf_pressure( P_input, PB_input )
THETA_input = wrf_theta( T_input )
print PRES_input.shape, THETA_input.shape,
```

#### Now we want to interpolate potential temperature onto pressure surfaces. This is done using wrf_to_pres in wrftools.interp. However, values of pressure must be monotonically increasing with height. Since pressure decreases with height, we just need to rearrange both arrays using array slicing.

```python
PRES_input = PRES_input[ :, ::-1, :, : ]
THETA_input = THETA_input[ :, ::-1, :, : ]

## specify the levels to interpolate to (in hPa)
interplevels = numpy.asarray( [ 1000, 950, 900, 850, 800, 750, 700, 650, 600, 550, 
                               500, 450, 400, 350, 300, 250, 200, 150, 100 ,50 ] )

THETA_out = wrf_to_pres( THETA_input, PRES_input, interplevels )
print THETA_out.shape
```

#### Now we're going to write the data out to the new netCDF file. As of right now, wrftools does not have any functions defined that take care of this automatically, so we must do it manually.

```python
## define variable dimensions.
level = outfile.createDimension( 'bottom_top', None )
time = outfile.createDimension( 'time', None )
lon = outfile.createDimension( 'south_north', 265 )
lat = outfile.createDimension( 'west_east', 265 )

## create XLONG and XLAT grid fields to write to the new file. 
## This is needed for plotting the data on a map
XLONG = outfile.createVariable( 'XLONG', 'f4', ('time', 'south_north', 'west_east') )
XLAT = outfile.createVariable( 'XLAT', 'f4', ('time', 'south_north', 'west_east') )

## copy the arrays of lons and lats to the outfile
XLONG[:] = infile.variables['XLONG'][:]
XLAT[:] = infile.variables['XLAT'][:]

## create variable THETA to write to, and then write to it.
THETA = outfile.createVariable( 'THETA', 'f4', ('time', 'bottom_top', 'south_north', 'west_east') )
THETA[:] = THETA_out[:]

## close the files
infile.close()
outfile.close()
```

#### And that is all there is to it. The data has successfully been written to our output file. It can be visualized using the package of your choice, but matplotlib/basemap is recommended. Below is a demonstration of the visualization.

```python
## initialize the basemap projection. This data happens to have a particular projection from the WRF model domain.
m = Basemap( projection='npstere', boundinglat=25, lon_0=0.231, resolution='l' )
## reopen out output file
presfile = Dataset('wrf_out.nc')
## we only need one time index of lats/lons since the grid does not move.
lons = presfile.variables['XLONG'][0]
lats = presfile.variables['XLAT'][0]
## we're plotting the 500mb level, which based on our interpolation levels we chose, is the 10th index of the array.
## we will also only be plotting the first time in the file.
theta = presfile.variables['THETA'][ 0, 10, :, : ]
print theta.shape
## draw the map boundaries
m.drawcountries()
m.drawcoastlines()
## plot the data using pcolormesh
cm = m.pcolormesh( lons, lats, theta, vmin=260, vmax=375, cmap=plt.get_cmap('spectral'), latlon=True )
cbar = plt.colorbar( cm )
plt.title( '500 hPa Potential Temperature \n Degrees Kelvin' )
presfile.close()
```
